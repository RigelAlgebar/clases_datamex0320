# Probs del dado

```python
def par(x):
    return x%2==0
```

```python
sample_space=[1,2,3,4,5,6]

caras=len(sample_space)

pares=len([e for e in sample_space if par(e)])

pares_prob=pares/caras

pares_prob
```

### binomial

```python
# https://es.wikipedia.org/wiki/Distribuci%C3%B3n_bin%C3%B3mica

from scipy.stats import binom
import matplotlib.pyplot as plt

# binom.pmf(k, n, p)   # k=nº de exitos, n=nº de repeticiones, p=probabilidad del evento

binom.pmf(3, 10, 0.5)  # ¿Cual es la prob de que salgan tres soles si tiro 10 veces la moneda?
```

```python
binom.pmf(5, 20, 1/6)  # ¿Cual es la prob de que salgan 5 doses si tiro 20 veces el dado?
```

```python
x=[i for i in range(20)]  # numero de exitos
y=[binom.pmf(x, 20, 0.5) for x in x] # su probabilidad

plt.plot(x, y);
sum(y)
```

```python
x=[i for i in range(20)]  # numero de exitos
y=[binom.cdf(x, 20, 0.8) for x in x] # su probabilidad

plt.plot(x, y);
```

### Poisson (ley de sucesos raros)

```python
# https://es.wikipedia.org/wiki/Distribuci%C3%B3n_de_Poisson
```

```python
from scipy.stats import poisson
```

```python
# poisson(x, lambda)   , poisson(evento, media)
```

```python
poisson.pmf(2, 1.4)
```

```python
x=[i for i in range(20)]  
y=[poisson.pmf(x, 1.4) for x in x] 

plt.plot(x, y);
```

```python
x=[i for i in range(20)]  
y=[poisson.pmf(x, 1.4) for x in x] 
plt.plot(x, y)
y=[poisson.pmf(x, 1) for x in x] 
plt.plot(x, y)
y=[poisson.pmf(x, 2.4) for x in x] 
plt.plot(x, y)
y=[poisson.pmf(x, 5) for x in x] 
plt.plot(x, y)
y=[poisson.pmf(x, 10) for x in x] 
plt.plot(x, y)
plt.show();
```

```python
x=[i for i in range(20)]  
y=[poisson.cdf(x, 1.4) for x in x] 

plt.plot(x, y);
```

```python
import numpy as np
from math import factorial


def poisson_custom(x, lamb):  # pmf
    
    return np.exp(-lamb)*lamb**x/factorial(x)
```

```python
poisson.pmf(2, 1.4)
```

```python
poisson_custom(2, 1.4)
```

### futbol

```python
import numpy as np
import seaborn as sns
```

```python
equipo_A=2.1  # medias de goles de dos equipos
equipo_B=1.4
```

```python
A=[poisson.pmf(i, equipo_A) for i in range(5)] 
B=[poisson.pmf(i, equipo_B) for i in range(5)] 
```

```python
A
```

```python
B
```

```python
M=np.outer(A, B)
```

```python
M
```

```python
import pandas as pd
pd.DataFrame(M)
```

```python
sns.heatmap(M)
```

```python
uno=np.sum(np.tril(M, -1))

empate=np.sum(np.diag(M))

dos=np.sum(np.triu(M, 1))
```

```python
print (uno, empate, dos)
```

```python
uno+empate+dos
```
